```python
coordinates_list = [
    (40.785091, -73.968285), (38.889484, -77.035278), (32.7318, -117.1976),
    (38.8913, -77.0300), (40.7580, -73.9855), (40.7587, -73.9787),
    (43.0828, -79.0742), (37.7694, -122.4862), (28.4177, -81.5812),
    (40.7021, -73.9973), (36.0160, -114.7377), (40.7480, -74.0048),
    (36.1129, -115.1765), (38.8780, -104.8618), (40.7794, -73.9632),
    (37.2982, -113.0263), (29.4241, -98.4936), (28.4444, -81.4670),
    (37.8199, -122.4783), (40.7061, -73.9968)
]

df["Coordinates"] = coordinates_list
```

```python
def color_by_rank(rank):
    if 1 <= rank <= 5:
        return "green"
    elif 6 <= rank <= 10:
        return "blue"
    elif 11 <= rank <= 15:
        return "red"
    else:
        return "magenta"
```

```python
# Initialize map centered on the U.S.
usa_map = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Add markers with popups
for idx, row in df.iterrows():
    popup_content = f"""
    <b>{row['Location']}</b><br>
    Rank: {row['Rank']}<br>
    City/State: {row['City/State']}<br>
    Annual Visitors: {row['Visitors (Annual)']}<br>
    Attractions: {row['Key Attractions']}
    """

    folium.Marker(
        location=row["Coordinates"],
        popup=folium.Popup(popup_content, max_width=300),
        icon=folium.Icon(color=color_by_rank(row["Rank"]), icon="info-sign")
    ).add_to(usa_map)

# Display the map
usa_map
```
